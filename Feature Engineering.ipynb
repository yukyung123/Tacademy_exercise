### 결측치 처리

삭제할 것인지, 대치할 것인지

```python
train[(train['workclass'] == '?')&(train['occupation'] == '?')]
#test데이터에서는 다를 수도 있음 
#income을 보면 확인, ?는 never worked라고 예측
```

```python
train.native_country.value_counts()
#'?'에러?,미기입?
```

```python
train.loc[(train.age >= 30)&(train.workclass=='?'), 'workclass'] = 'No'
#age가 30이 넘고, workclass의 값이 ?인 변수는 workclass의 값을 'No'로 정의
```
