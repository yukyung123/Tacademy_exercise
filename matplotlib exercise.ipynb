```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x= np.linspace(0,5,11)
x
```

```python
y=x**2
y
```

```python
#functional Method
plt.plot(x,y)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('tible')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(y,x)
```

```python
#object-oriented method
fig=plt.figure()
```

```python
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
ax1.plot(x,y)
ax2.plot(y,x)

ax1.set_title('larger plot')
ax2.set_title('smaller plot')
fig
```

```python
fig =plt.figure(figsize=(8,2),dpi =60)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
fig
```

```python
fig.savefig('my_pic.png',dpi=200)#png,jpg,,,dpi해상도#그림저장
```

```python
fig =plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x squared',color='purple')
ax.plot(x,x**3,label='x cubed')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes( [0, 0, 1, 1] )

ax.plot( x, y, color = 'purple', lw = 2, ls = '--' )

ax.set_xlim( [0, 1] )
ax.set_ylim( [0, 2] )
```
